```python
import os
import numpy as np
import cv2
import rasterio
from sklearn.model_selection import train_test_split
from tensorflow import keras
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, UpSampling2D, concatenate
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.models import Model
from tensorflow.keras.metrics import MeanIoU
```

```python
# Specify the paths to the image and label folders
image_folder = "/kaggle/working/sneha"
label_folder = "/kaggle/working/snefill"

# List all files in the image folder
image_files = [f for f in os.listdir(image_folder) if f.endswith('.tif')]

# Load and preprocess images and labels
images = []
labels = []
```

```python
# Function to load images using Rasterio
def load_image(image_path):
    with rasterio.open(image_path) as src_image:
        return src_image.read().transpose(1,2,0)
def load_label(label_path):
    with rasterio.open(label_path) as src_image:
        return src_image.read().transpose(1,2,0)
    
    
for image_file in image_files:
    image_path = os.path.join(image_folder, image_file)
    label_path = os.path.join(label_folder, image_file.replace('.tif', '.tiff'))

    image = load_image(image_path)
    label = load_label(label_path)

    images.append(image)
    labels.append(label)

images = np.array(images)
labels = np.array(labels)
```

```python
import tensorflow as tf

# Define a convolutional block
def Conv2dBlock(inputTensor, numFilters, kernelSize=3, doBatchNorm=True):
    x = tf.keras.layers.Conv2D(filters=numFilters, kernel_size=(kernelSize, kernelSize),
                               kernel_initializer='he_normal', padding='same')(inputTensor)

    if doBatchNorm:
        x = tf.keras.layers.BatchNormalization()(x)

    x = tf.keras.layers.Activation('relu')(x)

    x = tf.keras.layers.Conv2D(filters=numFilters, kernel_size=(kernelSize, kernelSize),
                               kernel_initializer='he_normal', padding='same')(x)

    if doBatchNorm:
        x = tf.keras.layers.BatchNormalization()(x)

    x = tf.keras.layers.Activation('relu')(x)

    return x

# Define the simplified UNet model
def SimplifiedUnet(inputImage, numFilters=64, dropout=0.1, doBatchNorm=True):
    c1 = Conv2dBlock(inputImage, numFilters * 1, kernelSize=3, doBatchNorm=doBatchNorm)
    p1 = tf.keras.layers.MaxPooling2D((2, 2))(c1)

    c2 = Conv2dBlock(p1, numFilters * 2, kernelSize=3, doBatchNorm=doBatchNorm)
    p2 = tf.keras.layers.MaxPooling2D((2, 2))(c2)

    c3 = Conv2dBlock(p2, numFilters * 4, kernelSize=3, doBatchNorm=doBatchNorm)
    p3 = tf.keras.layers.MaxPooling2D((2, 2))(c3)

    c4 = Conv2dBlock(p3, numFilters * 8, kernelSize=3, doBatchNorm=doBatchNorm)
    p4 = tf.keras.layers.MaxPooling2D((2, 2))(c4)

    c5 = Conv2dBlock(p4, numFilters * 16, kernelSize=3, doBatchNorm=doBatchNorm)

    u6 = tf.keras.layers.Conv2DTranspose(numFilters * 8, (3, 3), strides=(2, 2), padding='same')(c5)
    u6 = tf.keras.layers.concatenate([u6, c4])
    c6 = Conv2dBlock(u6, numFilters * 8, kernelSize=3, doBatchNorm=doBatchNorm)

    u7 = tf.keras.layers.Conv2DTranspose(numFilters * 4, (3, 3), strides=(2, 2), padding='same')(c6)
    u7 = tf.keras.layers.concatenate([u7, c3])
    c7 = Conv2dBlock(u7, numFilters * 4, kernelSize=3, doBatchNorm=doBatchNorm)

    u8 = tf.keras.layers.Conv2DTranspose(numFilters * 2, (3, 3), strides=(2, 2), padding='same')(c7)
    u8 = tf.keras.layers.concatenate([u8, c2])
    c8 = Conv2dBlock(u8, numFilters * 2, kernelSize=3, doBatchNorm=doBatchNorm)

    u9 = tf.keras.layers.Conv2DTranspose(numFilters * 1, (3, 3), strides=(2, 2), padding='same')(c8)
    u9 = tf.keras.layers.concatenate([u9, c1])
    c9 = Conv2dBlock(u9, numFilters * 1, kernelSize=3, doBatchNorm=doBatchNorm)

    output = tf.keras.layers.Conv2D(1, (1, 1), activation='sigmoid')(c9)
    model = tf.keras.Model(inputs=[inputImage], outputs=[output])
    return model

# Create the simplified UNet model with input shape (224, 224, 1)
inputs = tf.keras.layers.Input((224, 224, 3))
simplified_unet = SimplifiedUnet(inputs, dropout=0.07)

# Compile the model (you can use your preferred optimizer and loss function)
simplified_unet.compile(optimizer='Adam', loss='binary_crossentropy',metrics=['accuracy'])


```

```python
history=simplified_unet.fit(X_train,y_train,validation_data=(X_val,y_val),batch_size=2,epochs=100)
```

```python
import rasterio

# Define the path to your input image (to get the dimensions)
input_image_path = "/kaggle/working/snelabel/reshaped_LC82310932016065LGN00.tiff"

# Define the path to the output raster
output_raster_path = 'triaalraster.tif'

# Open the input image using Rasterio
with rasterio.open(input_image_path) as src:
    # Read the input image data
    input_data = src.read(1)  # Assuming it's a single-band image

    # Get the nodata value from the source image
    nodata_value = src.nodata

    # Create a new raster file with the same profile as the input image
    with rasterio.open(output_raster_path, 'w', **src.profile) as dst:
        # Replace nodata pixels with 0, otherwise keep the original values
        output_data = input_data.copy()
        output_data[input_data == nodata_value] = 0

        # Write the data to the new raster band
        dst.write(output_data, 1)

print(f"Raster with nodata pixels replaced by 0 saved as {output_raster_path}")
```
